```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch.nn.functional as F
import joblib


df = df_cleaned.copy()


df = df.fillna('Unknown')


X = df['name'].tolist()
y = df['use0'].tolist()


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

joblib.dump(label_encoder, "label_encoder.pkl")

#
label_encoder = joblib.load("label_encoder.pkl")


tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
class MedicineDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=64):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]

        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].squeeze(),
            'attention_mask': encoding['attention_mask'].squeeze(),
            'label': torch.tensor(label, dtype=torch.long)  }
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

train_dataset = MedicineDataset(X_train, y_train, tokenizer)
test_dataset = MedicineDataset(X_test, y_test, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(label_encoder.classes_))
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
```

```python
import torch

if torch.cuda.is_available():
    print(f"GPU is enabled: {torch.cuda.get_device_name(0)}")
else:
    print("GPU is not enabled.")
```

```python
from tqdm import tqdm

def train_model(model, train_loader, optimizer, epochs=3):
    model.train()
    for epoch in range(epochs):
        loop = tqdm(train_loader, desc=f'Epoch {epoch + 1}/{epochs}', leave=True)

        for batch in loop:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)

            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step()

            loop.set_postfix(loss=loss.item())

train_model(model, train_loader, optimizer)
```

```python
def predict_use(medicine_name):
    model.eval()
    with torch.no_grad():
        encoding = tokenizer.encode_plus(
            medicine_name,
            add_special_tokens=True,
            max_length=64,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        input_ids = encoding['input_ids'].to(device)
        attention_mask = encoding['attention_mask'].to(device)
        outputs = model(input_ids, attention_mask=attention_mask)
        prediction = torch.argmax(F.softmax(outputs.logits, dim=1), dim=1)
        return label_encoder.inverse_transform(prediction.cpu().numpy())[0]


```

```python

medicine_name = input("Enter Medicine Name: ")
print(f"Predicted Use: {predict_use(medicine_name)}")
```
