```python
import pandas as pd
```

```python
df = pd.read_csv("Home Remedies.csv")
df.head(20)
```

```python
df.head(5)
```

```python

df['Health Issue'] = df['Health Issue'].str.strip()


def get_remedies_by_disease(disease_name):

    filtered = df[df['Health Issue'].str.contains(disease_name, case=False, na=False)]
    
    if filtered.empty:
        print(f"No remedies found for disease: {disease_name}")
    else:
        print(f"\nHome remedies for '{disease_name.title()}':")
        for i, remedy in enumerate(filtered['Home Remedy'], 1):
            print(f"{i}. {remedy}")


```

```python
disease_input = input("Enter the name of the disease:").strip()
get_remedies_by_disease(disease_input)
```
