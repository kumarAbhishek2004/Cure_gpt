```python

selected_columns = ['name', 'sideEffect1', 'sideEffect2']
selected_columns = [col for col in selected_columns if col in df.columns]
df_selected = df[selected_columns].fillna("Not Available")
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_colwidth', 200)
print(df_selected.head(10))
```

```python
df_cleaned = df_selected.drop_duplicates(subset='name', keep='first')
df_cleaned.head(5)
```

```python
def combine_side_effects(row):
    side_effects = {str(row[col]).strip() for col in ['sideEffect1', 'sideEffect2']
                     if str(row[col]).strip().lower() != 'not available'}
    return ', '.join(side_effects) if side_effects else 'No side effects available'

df_cleaned['sideEffects'] = df_cleaned.apply(combine_side_effects, axis=1)
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch.nn.functional as F
import joblib
```

```python
df = df_cleaned_final.copy()


df = df.fillna('Unknown')


X = df['name'].tolist()
y = df['sideEffects'].tolist()


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
label_encoder_path = "/content/label_encoder.pkl"
joblib.dump(label_encoder, label_encoder_path)
print("LabelEncoder saved successfully!")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
class MedicineDataset(Dataset):
    def __init__(self, df):
        self.names = df['name'].tolist()
        self.sideEffects = df['sideEffects'].tolist()
        self.label_map = {effect: idx for idx, effect in enumerate(set(self.sideEffects))}

    def __len__(self):
        return len(self.names)

    def __getitem__(self, idx):
        encoding = tokenizer(self.names[idx], padding='max_length', truncation=True, max_length=64, return_tensors="pt")
        label = torch.tensor(self.label_map[self.sideEffects[idx]], dtype=torch.long)

        return {
            'input_ids': encoding['input_ids'].squeeze(0),
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'labels': label
        }

```

```python
from torch.optim import AdamW  # Use this instead
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# Load tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Ensure num_labels is correct (count of unique labels)
num_labels = len(set(df['sideEffects']))

# Load model with correct num_labels
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_labels).to(device)

# Define optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)


```

```python

dataset = MedicineDataset(df_cleaned_final)
train_loader = DataLoader(dataset, batch_size=8, shuffle=True)
```

```python
from tqdm import tqdm

def train_model(model, train_loader, optimizer, epochs=3):
    model.train()
    for epoch in range(epochs):
        loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=True)
        for batch in loop:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step()


            loop.set_postfix(loss=loss.item())

print("Starting Training...")
train_model(model, train_loader, optimizer, epochs=3)
print("Training complete!")


```

```python
def predict_side_effects(medicine_name):
    inputs = tokenizer(medicine_name, return_tensors="pt", padding="max_length", truncation=True, max_length=64).to(device)
    with torch.no_grad():
        outputs = model(**inputs)

    predicted_label_id = torch.argmax(outputs.logits, dim=1).cpu().numpy()[0]
    return dataset.sideEffects[predicted_label_id]


medicine_name = input("Enter medicine name: ")
print("Predicted Side Effects:", predict_side_effects(medicine_name))
```
